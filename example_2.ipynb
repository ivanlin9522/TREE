```python
import pandas as pd
import numpy as np
from prepare import *
from gurobipy import *
from MIP_solver import *
from MIP_solver_2 import *
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#y as a function of x
def fun_2(x1,x2):
    return (x1-5)**3+(x2-4.5)**3+(x1-1)**2+x2**2+500
```

```python
#generate x
x1=np.linspace(0, 5, 21)
x2=np.linspace(0, 5, 21)
```

```python
X1=np.zeros(len(x1)*len(x2))
X2=np.zeros(len(x1)*len(x2))
for i in range(len(x1)):
    for j in range(len(x2)):
        X1[i*len(x2)+j]=x1[i]
        X2[i*len(x2)+j]=x2[j]
x=np.column_stack((X1,X2))
y=fun_2(X1,X2)
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
x1_plot,x2_plot=np.meshgrid(x1, x2)
zs = np.array([fun_2(x_given,y_given) for x_given,y_given in zip(np.ravel(x1_plot), np.ravel(x2_plot))])
y_plot=zs.reshape(x1_plot.shape)
ax.plot_surface(x1_plot, x2_plot, y_plot)
plt.show() 
```

```python
flag=1 #regression tree
rf=get_rf(x,y,3,flag)
trees=list()
trees=get_input(rf)
```
